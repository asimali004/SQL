```python
from google.cloud import bigquery
```

```python
client = bigquery.Client()
```

```python
datasetref = client.dataset('chicago_crime', project="bigquery-public-data")
dataset    = client.get_dataset(datasetref)
tables     = client.list_tables(dataset)
```

```python
tableref = datasetref.table('crime')
table    = client.get_table(tableref)
```

```python
table
```

```python
table.schema
```

```python
client.list_rows(table,max_results = 12).to_dataframe()
```

```python
q1 = """
        SELECT distinct location_description
        FROM `bigquery-public-data.chicago_crime.crime`
        """
```

```python
client.query(q1).to_dataframe()
```

```python
client.list_rows(table,max_results=2).to_dataframe()
```

```python
q2 = """
        SELECT arrest ,count(arrest)
        FROM `bigquery-public-data.chicago_crime.crime`
        group by arrest
        """
```

```python
client.query(q2).to_dataframe()
```

```python
q3 = """
        SELECT arrest,year
        FROM `bigquery-public-data.chicago_crime.crime`
        """
```

```python
df = client.query(q3).to_dataframe()
```

```python
df.head()
```

```python
df.year.value_counts()
```

```python
dry_run_config = bigquery.QueryJobConfig(dry_run=True)
dry_run_query_job = client.query(q2, job_config=dry_run_config)
print("This query will process {} bytes.".format(dry_run_query_job.total_bytes_processed))
```

```python
ONE_MB = 1000*1000
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=ONE_MB)
safe_query_job = client.query(q3, job_config=safe_config)
safe_query_job.to_dataframe()
```

```python
ONE_GB = 1000*1000*1000
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=ONE_GB)
safe_query_job = client.query(q3, job_config=safe_config)
safe_query_job.to_dataframe()
```
